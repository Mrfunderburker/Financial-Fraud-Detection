#    Data cleaning, wrangling & pre-processing 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/Users/sa10/Downloads/PS_20174392719_1491204439457_log.csv')
```

```python
# drop unwanted columns
new_data = df.drop(columns=['nameDest', 'nameOrig', 'isFlaggedFraud'])
# apply one-hot encoding
new_data = pd.get_dummies(new_data, columns=['type'], dtype=int, drop_first=True)
# display info
new_data.info()
```

```python
num_cols = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
scaler = StandardScaler()
# apply the scaling transformation
new_data[num_cols] = scaler.fit_transform(new_data[num_cols])
# display
new_data[num_cols]
```

```python
# save to new csv
new_data.to_csv("new_data.csv")
```
